```python
import copy

class Algorithm:
    def __init__(self, nums, length):
        self.nums = nums
        self.length = length

    def insertion_sort(self):

        array = copy.deepcopy(self.nums)

        for index in range(1, self.length):
            key = array[index]
            prev = index - 1
            print("-------------------------")
            print(f"Check index: {index} with value {key}")

            print(f"Check previous index: {prev} with value {array[prev]}")

            while (prev+1) > 0 and array[prev] > key:
                print(f"Previous value: {array[prev]} greater than {key}")
                print(f"Shifting previous value to current index")
                array[prev+1] = array[prev] 
                print(f"New list: {array}")
                prev = prev -1

            print(f"Inserting key {key} to current index {prev+1}")
            array[prev+1] = key
            print(f"Final list: {array}")
    
    def selection_sort(self):

        array = copy.deepcopy(self.nums)

        for index in range(self.length):
            print(f"Current item {array[index]}, index {index}")
            min = index
            print(f"Min value: {array[min]}")

            for next in range(index+1, self.length):
                if array[next] < array[min]:
                    min = next

            print(f"New min: {array[min]}")
            print(f"Swapping A[{min}] with A[{index}]")

            array[index], array[min] = array[min], array[index]

            print(f"New list: {array}")


        print(f"Final list: {array}")


        
    def bubble_sort(self):

        array = copy.deepcopy(self.nums)

        for k in range(1, self.length):
            print("-------------------")
            print(f"Pass: {k}")
            for index in range(self.length-1):
                print(f"Pair: {array[index]}, {array[index+1]}")
                if array[index] > array[index+1]:
                    array[index], array[index+1] = array[index+1], array[index]
                print(f"New list: {array}")

        print(array)
```

```python
A = [2, 7, 4, 1, 5, 3]
n = len(A)
alg = Algorithm(A, n)
```

```python
print("Insertion Sort")
alg.insertion_sort()
```

```python
print("Selection Sort")
alg.selection_sort()
```

```python
print("Bubble Sort")
alg.bubble_sort()
```
